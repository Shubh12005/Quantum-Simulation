Norm calculation


```python
square_0 = np.square(bit_0)
sum_sqaures_0 = np.sum(square_0)
```

```python
norm_0 = np.sqrt(sum_sqaures_0)
```

FAIR COIN FLIP OPERARTOR

```python

c_fair = np.array([[0.5,0.5],[0.5,0.5]])
print(c_fair)
```

```python
state_0 = np.array([[1],[0]])
next_state = c_fair @ state_0
print(f'current state: {state_0} \n next state: {next_state}')
```

biased coin operator


```python
c_biased= ([[0.8,0.8],[0.2,0.2]])
next_state = c_biased @state_0
print(f'current state: {state_0}\n  next state : {next_state}')
```

stocastic matrix


```python
s_matrix = np.array([[0.9,0.1],[0.3,0.7]])
```

```python
row_sum = np.sum(s_matrix,axis =1)
print(f' row sum : { row_sum}')
```

```python
p_current= np.array([0.6,0.4])
p_next = p_current @ s_matrix
print(f'current state : {p_current} \n next state : {p_next}')
```

```python
# multiple times transition
```

```python
P= s_matrix
```

```python
p2 = P@P
print(f'p2: {p2}')
```

```python
p3 = P@P@P
print(f'p3: {p3}')
```
