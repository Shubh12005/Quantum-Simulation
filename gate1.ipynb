```
from qiskit import QuantumCircuit
from qiskit_aer import Aer
```

```
qc = QuantumCircuit(2, 2) # 2 qbits and classical bits
```

```
qc.x(0) # X (not)gate ,flips the qubit state ,No superpostion
```

```
qc.h(1)  #hadamard gate basically qubit q is now both 0 and 1 , exist in two states
```

```
qc.cx(1, 0) # C(not) gate , creating entanglement , applying to superpostion
```

```
qc.measure([0, 1], [0, 1]) # measuring the qubits, sys become either 01 , 10
```

```
simulator = Aer.get_backend('aer_simulator')
```

```
shot =10000
result = simulator.run(qc, shots=shot).result() # running the qc
```

```
counts = result.get_counts() # getting measurement counts
```

```
print("Measurement Results (Counts & Probabilities):\n")

for state in sorted(counts):
    count = counts[state]
    probability = count / shot
    print(f"State {state} = Count: {count}, Probability: {probability}")
```
