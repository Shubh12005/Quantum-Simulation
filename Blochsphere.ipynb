```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)

state = Statevector.from_instruction(qc)

probabilities = state.probabilities()
labels = ['|0⟩', '|1⟩']

plt.bar(labels, probabilities)
plt.ylabel("Probability")
plt.title("Statevector Probability Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Create circuit with Hadamard
qc = QuantumCircuit(1)
qc.h(0)

state = Statevector.from_instruction(qc)

alpha = state.data[0]
beta = state.data[1]

# Bloch coordinates
x = 2 * np.real(np.conj(alpha) * beta)
y = 2 * np.imag(np.conj(alpha) * beta)
z = np.abs(alpha)**2 - np.abs(beta)**2

# Plot Bloch sphere
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Sphere
u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:50j]
xs = np.cos(u)*np.sin(v)
ys = np.sin(u)*np.sin(v)
zs = np.cos(v)
ax.plot_surface(xs, ys, zs, alpha=0.2)

# State vector
ax.quiver(0, 0, 0, x, y, z, color='red', linewidth=3)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Bloch Sphere - Hadamard Gate")

plt.show()
```

```python


# Create 2-qubit circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

state = Statevector.from_instruction(qc)

probs = state.probabilities()
labels = ['|00⟩', '|01⟩', '|10⟩', '|11⟩']

plt.bar(labels, probs)
plt.ylabel("Probability")
plt.title("CNOT Gate Result (Entangled State)")
plt.show()
```
